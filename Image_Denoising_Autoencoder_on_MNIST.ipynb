```python
# ============================
# Advanced Image Denoising Autoencoder Project
# ============================

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks, optimizers
import tensorflow.keras.backend as K

# For hyperparameter tuning
!pip install -q keras-tuner
import kerastuner as kt

# For latent space visualization
from sklearn.manifold import TSNE

# ----------------------------
# 1. Data Preparation
# ----------------------------
# Load MNIST dataset (handwritten digits)
(x_train, _), (x_test, _) = tf.keras.datasets.mnist.load_data()

# Normalize images to [0,1] and reshape to include channel dimension
x_train = x_train.astype('float32') / 255.
x_test  = x_test.astype('float32')  / 255.
x_train = np.reshape(x_train, (-1, 28, 28, 1))
x_test  = np.reshape(x_test,  (-1, 28, 28, 1))

# ----------------------------
# 2. Add Noise to Data
# ----------------------------
# Define a noise factor to simulate a noisy environment.
noise_factor = 0.5
x_train_noisy = x_train + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_train.shape)
x_test_noisy  = x_test  + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_test.shape)
x_train_noisy = np.clip(x_train_noisy, 0., 1.)
x_test_noisy  = np.clip(x_test_noisy, 0., 1.)

# ----------------------------
# 3. Define Model Building Function for Tuning
# ----------------------------
def build_autoencoder(hp):
    """
    Build a deep convolutional autoencoder model.
    Hyperparameters (hp) are tuned using Keras Tuner.
    """
    input_img = layers.Input(shape=(28, 28, 1), name='input_image')

    # Encoder: Increase depth, use dropout and batch normalization
    x = layers.Conv2D(
        filters=hp.Int('enc_filters1', min_value=16, max_value=64, step=16),
        kernel_size=(3,3), activation='relu', padding='same', name='enc_conv1'
    )(input_img)
    x = layers.BatchNormalization(name='enc_bn1')(x)
    x = layers.MaxPooling2D((2,2), padding='same', name='enc_pool1')(x)
    x = layers.Dropout(rate=hp.Float('dropout1', 0.1, 0.3, step=0.1), name='enc_dropout1')(x)

    x = layers.Conv2D(
        filters=hp.Int('enc_filters2', min_value=16, max_value=64, step=16),
        kernel_size=(3,3), activation='relu', padding='same', name='enc_conv2'
    )(x)
    x = layers.BatchNormalization(name='enc_bn2')(x)
    encoded = layers.MaxPooling2D((2,2), padding='same', name='enc_pool2')(x)

    # Decoder: Mirror the encoder structure and use UpSampling
    x = layers.Conv2D(
        filters=hp.Int('dec_filters1', min_value=16, max_value=64, step=16),
        kernel_size=(3,3), activation='relu', padding='same', name='dec_conv1'
    )(encoded)
    x = layers.BatchNormalization(name='dec_bn1')(x)
    x = layers.UpSampling2D((2,2), name='dec_upsample1')(x)
    x = layers.Dropout(rate=hp.Float('dropout2', 0.1, 0.3, step=0.1), name='dec_dropout1')(x)

    x = layers.Conv2D(
        filters=hp.Int('dec_filters2', min_value=16, max_value=64, step=16),
        kernel_size=(3,3), activation='relu', padding='same', name='dec_conv2'
    )(x)
    x = layers.BatchNormalization(name='dec_bn2')(x)
    x = layers.UpSampling2D((2,2), name='dec_upsample2')(x)

    # Output layer with sigmoid activation
    decoded = layers.Conv2D(1, (3,3), activation='sigmoid', padding='same', name='dec_conv3')(x)

    # Construct and compile model
    autoencoder = models.Model(input_img, decoded, name='autoencoder_model')
    learning_rate = hp.Choice('learning_rate', values=[1e-3, 1e-4])
    autoencoder.compile(optimizer=optimizers.Adam(learning_rate=learning_rate),
                        loss='binary_crossentropy')
    return autoencoder

# ----------------------------
# 4. Hyperparameter Tuning with Keras Tuner
# ----------------------------
tuner = kt.RandomSearch(
    build_autoencoder,
    objective='val_loss',
    max_trials=5,
    executions_per_trial=1,
    directory='autoencoder_tuning',
    project_name='denoising_autoencoder'
)

print("Starting hyperparameter tuning...")
tuner.search(x_train_noisy, x_train,
             epochs=10,
             batch_size=128,
             validation_data=(x_test_noisy, x_test),
             callbacks=[callbacks.EarlyStopping(monitor='val_loss', patience=3)])
best_hp = tuner.get_best_hyperparameters(num_trials=1)[0]
print("Best Hyperparameters:")
print(best_hp.values)

# Get the best model and print summary
autoencoder = tuner.hypermodel.build(best_hp)
autoencoder.summary()

# ----------------------------
# 5. Train the Best Model with Callbacks
# ----------------------------
# Define callbacks: EarlyStopping and ModelCheckpoint to save best weights.
early_stop = callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
checkpoint = callbacks.ModelCheckpoint("best_autoencoder.h5", monitor='val_loss', save_best_only=True)
tensorboard_cb = callbacks.TensorBoard(log_dir="./logs", histogram_freq=1)

# Train the autoencoder model
history = autoencoder.fit(
    x_train_noisy, x_train,
    epochs=30,
    batch_size=128,
    shuffle=True,
    validation_data=(x_test_noisy, x_test),
    callbacks=[early_stop, checkpoint, tensorboard_cb]
)

# ----------------------------
# 6. Evaluate the Model
# ----------------------------
# Plot training & validation loss
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training vs Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

# Use the trained model to denoise test images
decoded_imgs = autoencoder.predict(x_test_noisy)

# Calculate the average reconstruction error on the test set
reconstruction_error = np.mean(np.square(x_test - decoded_imgs))
print(f"Average Reconstruction Error on Test Set: {reconstruction_error:.4f}")

# ----------------------------
# 7. Visualize the Latent Space with t-SNE
# ----------------------------
# Create an intermediate model to extract latent representations from the encoder
encoder_model = models.Model(autoencoder.input,
                             autoencoder.get_layer('enc_pool2').output)

# Get latent representations for test set
latent_reps = encoder_model.predict(x_test_noisy)
# Flatten latent representations for t-SNE (if necessary)
latent_flat = latent_reps.reshape(latent_reps.shape[0], -1)

# Use t-SNE to reduce to 2 dimensions
tsne = TSNE(n_components=2, random_state=42)
tsne_results = tsne.fit_transform(latent_flat)

plt.figure(figsize=(8, 6))
plt.scatter(tsne_results[:, 0], tsne_results[:, 1], s=2, cmap='viridis')
plt.title('t-SNE Projection of Latent Representations')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.show()

# ----------------------------
# 8. Display Original, Noisy, and Denoised Images
# ----------------------------
def display_images(original, noisy, decoded, n=10):
    """
    Display a grid of original, noisy, and denoised images.
    """
    plt.figure(figsize=(20, 8))
    for i in range(n):
        # Original images
        ax = plt.subplot(3, n, i + 1)
        plt.imshow(original[i].reshape(28, 28), cmap='gray')
        plt.title("Original")
        plt.axis('off')

        # Noisy images
        ax = plt.subplot(3, n, i + 1 + n)
        plt.imshow(noisy[i].reshape(28, 28), cmap='gray')
        plt.title("Noisy")
        plt.axis('off')

        # Denoised images
        ax = plt.subplot(3, n, i + 1 + 2*n)
        plt.imshow(decoded[i].reshape(28, 28), cmap='gray')
        plt.title("Denoised")
        plt.axis('off')
    plt.tight_layout()
    plt.show()

display_images(x_test, x_test_noisy, decoded_imgs)

# ----------------------------
# 9. Save the Model and Hyperparameters
# ----------------------------
# Save the final trained model and best hyperparameters for future use.
autoencoder.save("final_denoising_autoencoder.h5")
print("Model saved as 'final_denoising_autoencoder.h5'")
print("Best Hyperparameters:", best_hp.values)
```
